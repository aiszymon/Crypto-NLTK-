```python
import urllib.request
import json
```

```python
dates = ["2020-01-01", "2020-03-01", "2020-05-01", "2020-07-01", "2020-09-01", "2020-11-01", "2020-12-01"]

link = "https://data.messari.io/api/v1/assets/btc/metrics/price/time-series"

link += "?start=2020-01-01"
link += "&end=2021-01-01"
link += "&interval=1d"

url = urllib.request.urlopen(link)

data = url.read()
encoding = url.info().get_content_charset('utf-8')
```

```python
js = json.loads(data.decode(encoding))
j = json.dumps(js, indent=2)
j
```
